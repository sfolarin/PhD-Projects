```python

import numpy as np
import matplotlib.pyplot as plt
#import numdifftools as nd
from sympy.solvers import solve
from scipy.misc import derivative
from sympy import Symbol
import sympy
from sympy import solveset
from scipy import misc
#init_printing()
plt.rc('axes', linewidth=2)

Data = np.loadtxt('mcalldata.txt')

x = Data[:,0]


d =3

y = -Data[:,1]*1308
y1 = -Data[:,2]*1308
y2 = -Data[:,3]*1308
y3 = -Data[:,4]*1308
y4 = -Data[:,5]*1308
y5 = -Data[:,6]*1308
y6 = -Data[:,7]*1308
y7 = -Data[:,8]*1308
y8 = -Data[:,9]*1308
y9 = -Data[:,10]*1308
y10 = -Data[:,11]*1308
y11 = -Data[:,12]*1308
y12 = -Data[:,13]*1308
y13 = -Data[:,14]*1308
y14 = -Data[:,15]*1308
y15 = -Data[:,16]*1308
y16 = -Data[:,17]*1308
y17 = -Data[:,18]*1308
y18 = -Data[:,19]*1308

#y19 = -Data[:,20]
#y20 = -Data[:,21]



f = np.poly1d(np.polyfit(x, y, d))
f1 = np.poly1d(np.polyfit(x, y1, d))

f2 = np.poly1d(np.polyfit(x, y2, d))
f3 = np.poly1d(np.polyfit(x, y3, d))
f4 = np.poly1d(np.polyfit(x, y4, d))
f5 = np.poly1d(np.polyfit(x, y5, d))
f6 = np.poly1d(np.polyfit(x, y6, d))
f7 = np.poly1d(np.polyfit(x, y7, d))
f8 = np.poly1d(np.polyfit(x, y8, d))
f9 = np.poly1d(np.polyfit(x, y9, d))
f10 = np.poly1d(np.polyfit(x, y10, d))
f11 = np.poly1d(np.polyfit(x, y11, d))
f12 = np.poly1d(np.polyfit(x, y12, d))
f13 = np.poly1d(np.polyfit(x, y13, d))
f14 = np.poly1d(np.polyfit(x, y14, d))
f15 = np.poly1d(np.polyfit(x, y15, d))
f16 = np.poly1d(np.polyfit(x, y16, d))
f17 = np.poly1d(np.polyfit(x, y17, d))
f18 = np.poly1d(np.polyfit(x, y18, d))
#f19 = np.poly1d(np.polyfit(x, y19, d))
#f20 = np.poly1d(np.polyfit(x, y20, d))

p = [f,f1,f2,f3,f4,f5,f6,f7,f8,f9,f10,f11,f12,f13,f14,f15,f16,f17,f18]

# z = np.zeros((20,2))
# #print(z)
# temp = 310
#x = sympy.Symbol('x', integer=True)
dt = []
for i in p:
    #q = solveset(i)
    #print(q)
    #print(i)
    m = misc.derivative(i,0)
    dt.append(m)
     
#     temp = temp+10
    
    print(m)
    
    
    
    

np.savetxt('MC_fit.txt_0', dt)
plt.plot(x, y,  'og', lw=2 )
plt.plot(x, f(x), 'g', lw=2,label='320 K')

plt.plot(x, y1,  'ok', lw=2 )
plt.plot(x, f1(x), 'k', lw=2,label='330 K')

plt.plot(x, y2,  'om', lw=2 )
plt.plot(x, f2(x), 'm', lw=2,label='340 K')

plt.plot(x, y3,  'oy', lw=2 )
plt.plot(x, f3(x), 'y', lw=2,label='350 K')

plt.plot(x, y4,  'oc', lw=2 )
plt.plot(x, f4(x), 'c', lw=2,label='360K')

plt.plot(x, y5,  'or', lw=2 )
plt.plot(x, f5(x), 'r', lw=2,label='370 K')

plt.plot(x, y6,  'ob', lw=2 )
plt.plot(x, f6(x), 'b', lw=2,label='380 K')

plt.plot(x, y7,  'og', lw=2 )
plt.plot(x, f7(x), 'g', lw=2,label='390 K')

plt.plot(x, y8,  'oc', lw=2 )
plt.plot(x, f8(x), 'c', lw=2,label='400 K')

plt.plot(x, y9,  'oc', lw=2 )
plt.plot(x, f9(x), 'c', lw=2,label='410 K')

plt.plot(x, y10,  'ob', lw=2 )
plt.plot(x, f10(x), 'b', lw=2,label='420 K')

plt.plot(x, y11,  'oc', lw=2 )
plt.plot(x, f11(x), 'c', lw=2,label='430 K')

plt.plot(x, y12,  'oy', lw=2 )
plt.plot(x, f12(x), 'y', lw=2,label='440 K')


plt.plot(x, y13,  'oc', lw=2 )
plt.plot(x, f13(x), 'c', lw=2,label='450 K')


plt.plot(x, y14,  'om', lw=2 )
plt.plot(x, f14(x), 'm', lw=2,label='460 K')



plt.plot(x, y15,  'og', lw=2 )
plt.plot(x, f15(x), 'g', lw=2,label='470 K')


plt.plot(x, y16,  'oc', lw=2 )
plt.plot(x, f16(x), 'c', lw=2,label='480 K')



plt.plot(x, y17,  'ok', lw=2 )
plt.plot(x, f17(x), 'k', lw=2,label='490 K')


plt.plot(x, y18,  'oc', lw=2 )
plt.plot(x, f18(x), 'c', lw=2,label='500 K')



plt.legend(loc='upper left', shadow=True, fontsize='small')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(bbox_to_anchor=(1.0, 1.1))

plt.tick_params(direction='in', pad=10)
plt.tick_params('both', length=7, width=2)
#plt.title('My title',fontsize = 20)
plt.xlabel('E(kV/Cm)', fontsize = 20)
plt.ylabel('P($\mu C/cm^2$)',fontsize = 20)
plt.savefig('PE.eps', format='eps', dpi=300, bbox_inches='tight')
plt.savefig('MC_fit.jpeg', format='jpeg', dpi=300, bbox_inches='tight')

plt.show()

```
