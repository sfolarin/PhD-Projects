#### Matplotlib
Matplotlib, a potent and widely-used plotting library in Python, offers a diverse range of tools for crafting static, animated, and interactive visualizations. Numerous types of plots are achievable with Matplotlib. In a previous tutorial, we examined a simple line plot and scatter plot. See [Documentation](https://matplotlib.org/).


##### Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

##### Example

Let's create a simple line plot using a custom function

```python
import numpy as np

# Define the function
def func(x):
    return (x - 3) * (x - 5) * (x - 7) + 100

# Define the x range and calculate y
x = np.linspace(0, 10)
y = func(x)
```

##### Creating a figure & plot in it

```python
# Define a figure space that will contain the plot
fig = plt.figure(figsize =(4, 3),dpi=150)
#create a plot
plt.plot(x, y, 'r', linewidth=5)
```

```python
fig = plt.figure(figsize = (10,5), dpi=150)
plt.plot(x,y)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5), dpi=150)
ax[0,0].plot(x,y)
ax[0,1].plot(x,y)
ax[1,0].plot(x,y)
ax[1,1].plot(x,y)
```

```python
fig = plt.subplots(figsize=(10,5), dpi=150)
plt.plot(x,y)
```

##### Adding Titles and Labels
You can add a title to your plot with the `title` function, and labels to the x and y axes with the `xlabel` and `ylabel` functions. Additionaly you can change size of the fonts.

```python
fig = plt.figure(figsize =(4, 3),dpi=150)
plt.plot(x, y, 'r', linewidth=2)
plt.title("Line Plot",size=20)
plt.xlabel("X",size=18)
plt.ylabel("f(x)",size=18)
```

##### Changing the Plot Style
Matplotlib allows you to alter the style of your plot. You can change the line style, line color, and add markers for the data points. The style is defined by a string of characters that define the color, marker and line type, in that order. For example, `'go-'` stands for green dots with a solid 

```python
fig = plt.figure(figsize =(4, 3),dpi=150)

# Plotting with specific plot styles
plt.plot(x, y, 'go-')

# Adding title and labels
plt.title("Line Plot",size=20)
plt.xlabel("X",size=18)
plt.ylabel("f(x)",size=18);
```

##### Saving the Plot to a File
You can save your plot to a file using the `savefig` function. This allows you to keep a hard copy of your plot saved in a location you specify. The argument `bbox_inches='tight'` ensures that the figure is saved without excessive whitespace. You can also set the `dpi` if you are saving in an image format. There are many file formats support

- PNG
- JPEG
- SVG
- PDF
- EPS
- TIFF
- BMP
- PGF (for LaTeX)

```python
plt.savefig('lineplot.jpeg',bbox_inches='tight',dpi=300)
```

## Histograms and Density Plots
Histograms and density plots are both used to visualize the distribution of data. A histogram is a bar graph that shows the number of data points in each range of values. A density plot is a smooth curve that shows the probability of a data point falling within a range of values.

We will use Seaborn's `.histplot` make the histograms.

#### Plotting


```python
fig = plt.figure(figsize =(4, 3),dpi=150)
sns.histplot(data=df, x="bill_length_mm",hue='sex') # Creates an overlapping 
```

You can stack the plots on top of each other

```python
fig = plt.figure(figsize =(4, 3),dpi=150)
sns.histplot(data=df, x="bill_length_mm",hue='sex', multiple="stack")
```

with different  number of bins

```python
fig = plt.figure(figsize =(4, 3),dpi=150)
sns.histplot(data=df, x="bill_length_mm",hue='sex', bins=10)
```

#### Kernel Density
Kernel density estimation (KDE) is a non-parametric method for estimating the probability density function of a continuous random variable. It does this by smoothing the data points with a kernel function, which is a bell-shaped curve. The kernel function is chosen to be symmetric and to have a peak at the mean of the data. 

We can use `kde` parameter with  `histplot` to get an estimation of KDE

```python
fig = plt.figure(figsize =(4, 3),dpi=150)
sns.histplot(data=df, x="bill_length_mm",hue='sex', kde=True)
```

#### 2D histograms

```python
fig = plt.figure(figsize =(4, 3),dpi=150)
sns.histplot(df, x="bill_depth_mm", y="body_mass_g", hue="sex")
```

There are multiple sets of parameters you can experiment with to customize your plot. Refer to the [documentation](https://seaborn.pydata.org/generated/seaborn.histplot.html) for additional details.

## Correlation Plots: Heatmap and Pairplots

Correlation plots and heatmaps are indispensable tools in data analysis, providing quick insights and aiding decision-making. 

`Correlation Plots`: High correlations between variables might indicate redundancy. In feature selection, you can identify and exclude highly correlated variables to simplify models and improve interpretability. Additionally, having too many correlated variables in your model may affect its performance.

`Heatmaps`: Heatmaps provide a color-coded matrix representation of data. They're particularly useful for visualizing large datasets or matrices. Heatmaps can reveal patterns, clusters, and variations within data that might not be evident in raw numbers. They can also provide a visual interpretation for your correlation matrix, helping you select the right set of features to increase your model's performance and interpretability.

`Pairplots`:Pairplots are visual tools for exploring relationships among numerical variables in a dataset. They show scatter plots between variable pairs and help identify correlations quickly. Useful for initial data insight, feature selection, and spotting trends.

Let's use a Seaborn Dataset to generate the correlation matrix using Pandas and visualize it via a heatmap.

#### Loading the dataset

We are going to use the `mpg` dataset. It contains information about various car models' fuel efficiency (miles per gallon) and other related attributes.

```python
df = sns.load_dataset("mpg")
df.head()
```

Since it is a Pandas dataframe, we can generate the correlation matrix of the numeric columns in the loaded dataframe.

```python
# Calculate the correlation matrix on the numeric columns
corr = df.select_dtypes('number').corr()

corr
```

As you can see, it is a correlation matrix with diagonal entries being 1 (self-correlation), and all the other entries showing the Pearson correlation coefficient ([read more](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient)). However, these are all scalar values, not too convenient to interpret.


#### Heatmap

Lets use the Seaborn's `heatmap` and see how it looks!

```python
fig = plt.figure(figsize =(5, 4),dpi=150)
sns.heatmap(corr, annot=True, cmap='BrBG',vmax=1,vmin=-1, cbar_kws={'label': 'Pearson correlation(r)'})
```

The `cmap` argument defines a colormap. There are many [options](https://seaborn.pydata.org/tutorial/color_palettes.html)  available. The `annot` keyword places the values on the heatmap itself. The `vmax` and `vmin` keywords define the limits of the colorbar values. You can also pass arguments for the color bar using the `cbar_kws` parameter. In this example, we used it to set the label of the colorbar axis. There are always some additional parameters to play around with. Please see the [documentation](https://seaborn.pydata.org/generated/seaborn.heatmap.html).

#### Pairplots

```python
sns.pairplot(df, hue="horsepower",height=1.5)
```

It displays the correlation in the form of a scatterplot. See [documentation](https://seaborn.pydata.org/generated/seaborn.pairplot.html) for information regarding the additional parameters.

## 3D Plots

3D plotting enriches data visualization by adding depth and complexity, aiding in understanding intricate relationships, fostering engagement, and enabling interactive exploration. These plots provide insights into data with an additional dimension, adding depth to your visualizations. While we have used Seaborn so far, since it does not provide the flexibility of 3D plotting, we will use Matplotlib instead




Let's create a dataset first

```python
import numpy as np

#This will generate data for plotting a 3D Torus


u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, 2 * np.pi, 100)
U, V = np.meshgrid(u, v)
a, b = 1, 0.2
x = (a + b * np.cos(V)) * np.cos(U)
y = (a + b * np.cos(V)) * np.sin(U)
z = b * np.sin(V)
```

since it is a 3D plot we need to add projeciton axis

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize =(4, 4),dpi=150)
ax = fig.add_subplot(111, projection='3d') #Adding the projection axis

ax.plot_surface(x, y, z, cmap='RdBu')

# Set labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```

There are many additional kinds of plots you can create with Matplotlib. Refer to the [documentation](https://matplotlib.org/stable/gallery/mplot3d/index.html) for more details.

## Best Practices in Data Visualization

When working with Matplotlib and Seaborn, adhering to best practices in data visualization enhances the clarity and impact of your visualizations. Here are some key points to consider:

`Choosing the Right Plot for the Data`: Select visualizations that effectively represent your data. Bar plots, scatter plots, and line plots have different strengths based on the data's nature.

`Color Choices and Accessibility`: Opt for colors that are visually appealing and easily distinguishable, considering colorblindness. Seaborn's color palettes offer great options.

`Simplifying and Removing Clutter`: Keep your visualizations clean by removing unnecessary grid lines, axes, or decorations. This reduces visual noise and emphasizes the data.

`Conveying Clear Messages with Visualizations`: Every visualization should tell a story. Use informative titles, labels, and annotations to convey the message you want to share.

You should be able to create compelling visualizations that effectively communicate insights, making your data more accessible and understandable to a wider audience.
